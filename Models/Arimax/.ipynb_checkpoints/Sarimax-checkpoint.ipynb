```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_file = "../TrainTestFiles/Train-Test-0"

trainDataset = pd.read_csv("{}/Train.csv".format(train_file), index_col=0)

sarimax = SARIMAX(
    endog=trainDataset["Cases"].values,
    exog=trainDataset["Searches"].values,
    order=(4,1,4),
    seasonal_order=[0,0,0,52],
)
res = sarimax.fit(disp=False)
print(res.summary())
```

```python
# In-sample one-step-ahead predictions
predict = res.get_prediction()
predict_ci = predict.conf_int()


predict_dy = res.get_prediction(dynamic=0)
predict_dy_ci = predict_dy.conf_int()
```

```python
# Graph
fig, ax = plt.subplots(figsize=(15,8))
npre = 4
ax.set(title='Cases Plot', xlabel='Date', ylabel='Cases')

# Plot data points
trainDataset.loc[:, 'Cases'].plot(ax=ax, style="o", label="Observed")

# # Plot predictions
plt.plot(predict.predicted_mean, "r--", label="One-step-ahead forecast")

predict_dy.predicted_mean.loc['1977-07-01':].plot(ax=ax, style='g', label='Dynamic forecast (1978)')
# ci = predict_dy_ci.loc['1977-07-01':]
# ax.fill_between(ci.index, ci.iloc[:,0], ci.iloc[:,1], color='g', alpha=0.1)

# legend = ax.legend(loc='lower right')

```
