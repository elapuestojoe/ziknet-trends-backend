```python
import numpy as np
import pandas as pd
import pyflux as pf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../../data/Mexico/processed_data/Veracruz_2016-2018.csv", index_col=0)
```

```python
plt.figure(figsize=(15,5))
plt.plot(data.index,data['Cases'], "ro")
plt.ylabel('Cases')
plt.title('Zika cases 2016-2018')
plt.plot();
```

```python
model = pf.ARIMAX(data=data, formula='Cases~Searches',
                  ar=1, ma=1, family=pf.Normal())
x = model.fit("MLE")
x.summary()
```

```python
model.plot_fit(figsize=(15,10))
```
